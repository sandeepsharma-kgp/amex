```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv("Training_Dataset.csv")
```

```python
df.head()
```

```python
df.drop(['mvar46','mvar47','mvar48'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
m = df[['mvar49','mvar50','mvar51']].as_matrix()
```

```python
def cat(m):
    l = []
    for i in range(len(m)):
        if(m[i][0] == 0 and m[i][1] == 0 and m[i][2] == 0):
            l.append("na")
            continue
        elif(m[i][0] == 0 and m[i][1] == 0 and m[i][2] == 1):
            l.append("Credit")
            continue
        elif(m[i][0] == 0 and m[i][1] == 1 and m[i][2] == 0):
            l.append("Elite")
            continue
        elif(m[i][0] == 1 and m[i][1] == 0 and m[i][2] == 0):
            l.append("Supp")
            continue
    return(l)
```

```python
categories = cat(m)
```

```python
df["classactual"] = Series(categories, index=df.index)
```

```python
df.drop(['mvar49','mvar50','mvar51'],axis = 1,inplace = True)
```

```python
ind_dummies = pd.get_dummies(df['mvar12'],prefix='indspend')
df = pd.concat([df,ind_dummies],axis=1)
```

```python
df2 = df.drop(['cm_key','mvar1','mvar12'],axis = 1)
```

```python
df2.drop(['classactual'],axis = 1,inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2_scaled = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)
```

```python
df2_scaled.head()
```

```python
corr = df2_scaled.corr()
```

```python
m = ~(corr.mask(np.eye(len(corr), dtype=bool)).abs() > 0.7).any()
```

```python
n = pd.DataFrame(m)
n.rename(columns={0: 'val'},inplace = True)
o = np.where(n['val']==True)
o
```

```python
p = n.ix[o[0]]
q = p.reset_index()
r = q['index']
```

```python
df2_new = df2_scaled.ix[:,r]
```

```python
df2_new.head()
```

```python
Y = df['classactual']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
X_train,X_test,Y_train,Y_test= train_test_split(df2_new,Y,test_size = 0.4,random_state = 3)
```

```python
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(X_train,Y_train)
Y_predict = knn.predict(X_test)
```

```python
print metrics.accuracy_score(Y_test,Y_predict)
```
