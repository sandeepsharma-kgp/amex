```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
```

```python
df = pd.read_csv("Training_Dataset.csv")
```

```python
df.head()
```

```python
df.drop(['mvar46','mvar47','mvar48'],axis = 1,inplace = True)
```

```python
df.head()
```

```python
m = df[['mvar49','mvar50','mvar51']].as_matrix()
```

```python
def cat(m):
    l = []
    for i in range(len(m)):
        if(m[i][0] == 0 and m[i][1] == 0 and m[i][2] == 0):
            l.append("na")
            continue
        elif(m[i][0] == 0 and m[i][1] == 0 and m[i][2] == 1):
            l.append("Credit")
            continue
        elif(m[i][0] == 0 and m[i][1] == 1 and m[i][2] == 0):
            l.append("Elite")
            continue
        elif(m[i][0] == 1 and m[i][1] == 0 and m[i][2] == 0):
            l.append("Supp")
            continue
    return(l)
```

```python
categories = cat(m)
```

```python
df["classactual"] = Series(categories, index=df.index)
```

```python
df.drop(['mvar49','mvar50','mvar51'],axis = 1,inplace = True)
```
