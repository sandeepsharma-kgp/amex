```python
dft2 = pd.read_csv("Training_Dataset.csv")
m = dft2[['mvar49','mvar50','mvar51']].as_matrix()

def cat(m):
    l = []
    for i in range(len(m)):
        if(m[i][0] == 0 and m[i][1] == 0 and m[i][2] == 0):
            l.append("na")
            continue
        elif(m[i][0] == 0 and m[i][1] == 0 and m[i][2] == 1):
            l.append("Credit")
            continue
        elif(m[i][0] == 0 and m[i][1] == 1 and m[i][2] == 0):
            l.append("Elite")
            continue
        elif(m[i][0] == 1 and m[i][1] == 0 and m[i][2] == 0):
            l.append("Supp")
            continue
    return(l)

categories = cat(m)

dft2["classactual"] = Series(categories, index=dft2.index)
```
